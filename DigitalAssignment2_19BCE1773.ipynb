###  Reading and tokenizing the text file

```python
textfile = open('corpora_1773.txt','r')
textfile.read()
```

```python
import nltk
file_content = open('corpora_1773.txt').read()
tokens = nltk.word_tokenize(file_content)
print(tokens)
```

### Word segmentation

### Sentence Segementation

```python
sentences = nltk.sent_tokenize(file_content)
```

```python
sentences
```

### Convert to lowercase

```python
lower_tokens = [token.lower() for token in tokens]
lower_tokens
```

### â€¢	Stop words removal

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
filtered_sentence = [w for w in tokens if not w.lower() in stop_words]
  
filtered_sentence = []
for w in lower_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)

print(filtered_sentence)
```

### Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

# getting the stem words from the list of filtered words
for w in filtered_sentence:
    print(w, " : ", ps.stem(w))
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

for w in filtered_sentence:
    print(w, " : ", lemmatizer.lemmatize(w))
```

### POS tagging

```python
for sent in filtered_sentence:
    print(nltk.pos_tag(nltk.word_tokenize(sent)))
```
